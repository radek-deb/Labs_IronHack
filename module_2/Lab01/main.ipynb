#### Use your previous code but now, resize your plot.

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
math.e
```

```python
# your code here
y1 = x**2
y2 = (math.e)**x
```

```python
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(10,5))

#First plot
ax1.plot(x,y1, 'b', linewidth=4)
ax1.plot(x,y2, 'g',linewidth=4)
ax1.set_title('y=x^2', fontsize=14)   
ax1.set_xlabel('x label', fontsize=12)     
ax1.set_ylabel('y label',  fontsize=12)   

#Second plot
ax2.plot(x,y1, 'b', linewidth=4)
ax2.plot(x,y2, 'g',linewidth=4)
ax2.set_title('y=e^x', fontsize=15)
ax2.set_xlabel('x label', fontsize=12)
ax2.set_ylabel('y label', fontsize=12)
ax2.set_yscale('log')

plt.suptitle("Comparison of 2 functions", fontsize=16)

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(10,5))

#First plot
ax1.plot(x,y1, 'b', linewidth=4, label='square')
ax1.plot(x,y2, 'g',linewidth=4, label='e to power of x')
ax1.set_title('y=x^2', fontsize=14)   
ax1.set_xlabel('x label', fontsize=12)     
ax1.set_ylabel('y label',  fontsize=12)
ax1.legend()   


#Second plot
ax2.plot(x,y1, 'b', linewidth=4, label = 'square')
ax2.plot(x,y2, 'g',linewidth=4, label = 'e to power of x')
ax2.set_title('y=e^x', fontsize=15)
ax2.set_xlabel('x label', fontsize=12)
ax2.set_ylabel('y label', fontsize=12)
ax2.set_yscale('log')
ax2.legend()


plt.suptitle("Comparison of 2 functions", fontsize=16)

plt.show()
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
fig = plt.figure()
sns.displot(titanic, x='Age_int')
mean=titanic['Age'].mean()
median=titanic['Age'].median()
std=titanic['Age'].std()
plt.vlines(mean, 0, mean, color='crimson',linewidth=4, label = 'mean')
plt.vlines(median, 0, median, color='green',linewidth=4, label = 'median')
plt.vlines(std, 0, std, color='c',linewidth=4, label = 'standard deviation')
plt.legend(loc=1, prop={'size':10})
plt.title('Age distribution with basic statistics')
plt.show()
```
