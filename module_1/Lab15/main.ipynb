Question 1: Find the questions and answers of last month.

```python
#To convert to seconds
from datetime import datetime

#get current date
dt = datetime.today()  
seconds_td = int(dt.timestamp())
seconds_td


```

```python
#date last month
t = datetime(2021, 11, 6, 0, 0)
seconds_lm = int(t.timestamp())
seconds_lm
```

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=seconds_lm, todate=seconds_td)
questions
```

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate=seconds_lm, todate=seconds_td)
answers
```

Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'.

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=int(datetime(2021,12,6,0,0).timestamp()), todate=seconds_td, tagged='python', sort='votes')
questions
```

```python
data = pd.DataFrame(questions['items'])
data.value_counts('score')
```
