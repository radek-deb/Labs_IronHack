#### Count the number of tweets by a given Twitter account.
Ask the user for the handle (@handle) of a twitter account. You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account.

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
#
#It was impossible to do regular web scraping because the wepsite was heavly using javascript
# So insted of usinf Beautiful Soup I use selenium 
# , which uses drivers to interact with the browsers and allow the javascript to run properly and do webscrpaping
#  
#
````

````python
# your code here
from selenium.webdriver.common.by import By
from selenium import webdriver
url = r'https://twitter.com/'
browser = webdriver.Chrome()
import time
````

````python

try:
	handle = input("Please enter your handle (@handle) of a twitter account: ").lstrip('@')
	browser.get(url+handle)
	# I had to add sleep finction because it took some time to acess the webpage 
	time.sleep(5)
	# Number of tweets
	# Is in the sibling div of H2 which holds name of Handle at the top of the page
	# Adele
	# 344 Tweets
	elem = browser.find_element(By.XPATH, "//h2/following-sibling::div")
	print("This account posted:")
	print(elem.text)
except:
	print("I didn't understand or Twitter account doesn't exist. Try again")
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
url = 'https://books.toscrape.com/catalogue/page-1.html'
htlm = requests.get(url).content
soup = BeautifulSoup(htlm,"lxml")
book = soup.find_all('h3')
````

````python
tittles = [pos.find('a')['title'] for pos in book]
tittles
````

````python
price = soup.find_all('p', class_='price_color')
prices = [cena.get_text() for cena in price]
prices
````

````python
stock = soup.find_all('p', class_='instock availability')
in_stock = [elem.get_text().strip('\n ') for elem in stock]
in_stock 
````

````python
# your code here
data = []
for num in range(1,51):
	url = f'https://books.toscrape.com/catalogue/page-{num}.html'
	htlm = requests.get(url).content
	soup = BeautifulSoup(htlm,"lxml")
	book = soup.find_all('h3')
	tittles = [pos.find('a')['title'] for pos in book]
	price = soup.find_all('p', class_='price_color')
	prices = [cena.get_text() for cena in price]
	stock = soup.find_all('p', class_='instock availability')
	in_stock = [elem.get_text().strip('\n ') for elem in stock]
	bokks_page = []
	for ind in range(len(tittles)):
		bokks_page.append([tittles[ind], prices[ind], in_stock[ind]])
	data.append(bokks_page) 
data

````
